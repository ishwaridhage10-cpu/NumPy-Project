```python
import numpy as np

array = np.random.randint(1, 101, (4, 4))

print("4x4 Random Integer Matrix:")
print(array)

print("\nShape:", array.shape)
print("Dimension (ndim):", array.ndim)
print("Total Number of Elements (size):", array.size)
print("Data Type:", array.dtype)
print("Minimum Value:", array.min())
print("Maximum Value:", array.max())
```
