```python
# Replace 0 with mean (excluding 0) in all numeric columns
for col in df.select_dtypes(include=["int64", "float64"]).columns:
    if (df[col] == 0).any():
        mean_value = df[df[col] != 0][col].mean()
        df[col] = df[col].replace(0, mean_value)

# Round numeric columns to 2 decimal places
df = df.round(2)

# Display first 5 rows
print(df.head())
```

```python
cols = ["age", "ejection_fraction", "serum_creatinine"]

plt.figure(figsize=(10,3))

for i, col in enumerate(cols, 1):
    plt.subplot(1,3,i)
    sns.histplot(df[col])
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
sns.countplot(x="anaemia", data=df)
plt.title("Count Plot of Anaemia")
plt.xlabel("Anaemia")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="age", y="serum_creatinine", data=df)
plt.title("Age vs Serum Creatinine")
plt.xlabel("Age")
plt.ylabel("Serum Creatinine")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
