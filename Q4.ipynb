```python
import numpy as np

array1 = np.linspace(0, 5, 10)
array2 = np.linspace(-10, 10, 15)

print("a) 10 equally spaced numbers between 0 and 5:")
print(array1)
print("Length:", len(array1))

print("\nb) 15 equally spaced numbers between -10 and 10:")
print(array2)
print("Length:", len(array2))
```
